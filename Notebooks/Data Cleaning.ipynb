## Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('joined_results.csv')
df.info()
```

```python
# Convert to appropriate data types

cols_to_convert = ['start_date','end_date','timestamp']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_datetime)
df.info()
```

```python
# Basic descriptive statistics

df.describe().round(2)

# Likely duplicated entries since event_id counts of 403,967 is more than its max number of 400,000.
```

```python
# Check number of NULL values

df.isnull().sum()
```

```python
# Checking for Outliers (keep due to valid users and no unreasonable user age value)

Q1 = df['user_age'].quantile(0.25)
Q3 = df['user_age'].quantile(0.75)
IQR = Q3-Q1

outliers = df[(df['user_age'] < Q1 - 1.5*IQR) | (df['user_age'] > Q3+1.5*IQR)]
print(outliers)
```

```python
# Check & Remove Partial Duplicates (3966 entries deleted due to conflicting ad entries with multiple users)

df_clean = df.drop_duplicates(subset=['event_id'])
df_clean.info()
```

```python
# Save clean data to csv file

df_clean.to_csv ('cleaned_joined_data.csv')   
```
